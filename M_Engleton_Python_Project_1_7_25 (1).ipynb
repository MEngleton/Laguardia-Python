```python
#dataframe to plot with only the mean algebra Regent scores for 3 years for the 2 chosen schools

algebra = df2.iloc[[130, 131, 132, 782, 783, 784]]
```

```python
algebra
```

```python
algebra.columns
```

```python
#Create a data frame that dropped all unnecessary columns and columns with too many "s" found above
algebra.drop(['School DBN', 'School Level', 'Regents Exam', 'Total Tested', 'Number Scoring Below 65', 'Percent Scoring Below 65', 'Number Scoring 65 or Above',
              'Percent Scoring 65 or Above', 'Number Scoring 80 or Above', 'Percent Scoring 80 or Above', 'Number Scoring CR',
              'Percent Scoring CR', 'Borough', 'District'], axis=1, inplace=True)
```

```python
algebra
```

```python
#Visualization
algebra_table = algebra.pivot_table(values="Mean Score", index="Year", columns="School Name").plot()

# set axis labels
algebra_table.set_xlabel('Year')
algebra_table.set_ylabel('Average Grade')

# set markers
markers = ['s', 'd']
for i, line in enumerate(algebra_table.get_lines()):
    line.set_marker(markers[i])

#set x axis tick mark labels
algebra_table.set_xticks((2015, 2016, 2017))
algebra_table.set_xticklabels(('$2015$', '$2016$', '$2017$'))
```

Above is a line graph comparing the average scores on the Common Core Algebra Regent Exam from 2015 to 2017 in 2 high schools in Manhattan: Marta Valle High School and NYC iSchool.
